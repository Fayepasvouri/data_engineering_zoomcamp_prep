```python
def shortestCompletingWord(license: str, words) -> str:
    
    from collections import Counter 

    license_count = Counter(ch for ch in license.lower() if ch.isalpha())
    
    nonvalid_words = []
    for word in words:
        word_counter = Counter(word)
    
        for key, val in license_count.items():
            if word_counter[key] < val:
                nonvalid_words.append(word)
    
    valid_words = []
    for word in words:
        if word not in nonvalid_words:
            valid_words.append(word)
    
    return min(valid_words, key=len)
    
shortestCompletingWord(license = "1s3 PSt", words = ["step","steps","stripe","stepple"])

```
