```python
# Stacks are implemeted like Lists in Python 
# LIFO - Last In First Out

# check if last character which was inserted is closing or opening bracket, 
# is bigger or smaller than next char etc and pop it from the stack 

# famous use case: balancing parentheses in an expression, find duplicate chars and remove them keeping lexicographical order

# stack = [], stack.append(), stack.pop(), stack[-1] to check last element
```

```python
# Easy Leetcode no. 20 Valid Parenthesis

from typing import List
def isValid(s: List[str]) -> bool:

    stack = []

    mydict = {')': '(', ']':  '[', '}':'{'}

    for key, val in mydict.items(): # for loop with pop and append has O(n) time complexity
        if val in s:
            stack.append(val)
        elif stack and stack.pop() != mydict[val]:
            return False
    return not stack


isValid(s = "()[]{}") # stack size grows with the imput size as we keep pushing so space complexity is O(n)
```
