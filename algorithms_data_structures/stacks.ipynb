```python
# Stacks are implemeted like Lists in Python 
# LIFO - Last In First Out

# check if last character which was inserted is closing or opening bracket, 
# is bigger or smaller than next char etc and pop it from the stack 

# famous use case: balancing parentheses in an expression, find duplicate chars and remove them keeping lexicographical order

# stack = [], stack.append(), stack.pop(), stack[-1] to check last element
```

```python
# Easy Leetcode no. 20 Valid Parenthesis

from typing import List
def isValid(s: List[str]) -> bool:

    stack = []

    mydict = {')': '(', ']':  '[', '}':'{'}

    for key, val in mydict.items(): # for loop with pop and append has O(n) time complexity (linear time)
        if val in s:
            stack.append(val)
        elif stack and stack.pop() != mydict[val]:
            return False
    return not stack


isValid(s = "()[]{}") # stack size grows with the imput size as we keep pushing so space complexity is O(n)
```

```python
# Medium level leetcode 316 Remove Duplicate Letters

from collections import Counter

def removeDuplicateLetters(s: str) -> str:
    stack = []
    seen = set()
    count = Counter(s)
        
    for char in s:
        count[char] -= 1
        if char in seen:
            continue
        
        while stack and char < stack[-1] and count[stack[-1]] > 0:
            seen.discard(stack.pop())
            
        stack.append(char)
        seen.add(char)
        
    return "".join(stack)

removeDuplicateLetters(s = "cbacdcbc")  # time complexity: O(n) space complexity: O(1) because the stack will contain at most 26 letters of the alphabet
```
