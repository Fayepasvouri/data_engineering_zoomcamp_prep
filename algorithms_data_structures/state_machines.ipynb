```python
def find_increasing_runs(nums):
    current = 0 
    best = 0 
    prev = None 

    for x in nums:
        if prev is None or x <= prev:
            current = 1
        else:
            current += 1
        best = max(best, current)
        prev = x
    return best

find_increasing_runs([1, 2, 2, 3, 1, 4, 5])

```
