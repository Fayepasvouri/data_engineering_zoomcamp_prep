### A greedy algorithm builds a solution step-by-step by always choosing the best immediate (local) option, hoping it leads to the global optimum.

Key idea:
➡️ At each step, pick the choice that looks the best right now.

Greedy works only when the problem has the greedy-choice property → meaning local optimum choices lead to global optimum.

Examples that greedy solves optimally:

- Fractional knapsack

- Activity selection

- Job sequencing

- Huffman coding

Not the 0-1 Knapsack. this is solved with Dynamic Programmic

```python
# uses tuples

def greedy_fractional_knapsack(items, capacity):

    # we need to sort the values
    items.sort(key = lambda x: x[0] / x[1], reverse=True)

    total_value = 0
    remaining_capacity = capacity

    for value, weight in items:
        if weight <= remaining_capacity:
            total_value += value
            remaining_capacity -= weight
        else:
            # create fraction
            fraction = remaining_capacity / weight
            total_value += value * fraction
            break

    return total_value

greedy_fractional_knapsack(items= [(30, 10), (40, 20), (60, 30)], capacity = 20)
```

```python
# easy leetcode example 455 assign cookies

def findContentChildren(g, s) -> int:
    
    g.sort()
    s.sort()
    
    i = j = 0
    count = 0
    
    while i < len(g) and j < len(s):
        if s[j] >= g[i]:
            count += 1
            i+=1
            j+=1
        else:
            j+=1
    return count
                
findContentChildren(g=[1,2,3], s=[1,1])
```

```python
# leetcode example 122 best time to buy and sell stock II
def maxProfit(prices) -> int:
        profit = 0
        for idx in range(len(prices)-1): # otherise out of index bounds due to idx+1
            if prices[idx+1] > prices[idx]:
                profit += prices[idx+1] - prices[idx] 
        return profit

maxProfit(prices = [7,1,5,3,6,4])
```

```python
def greedy_portfolio_allocation(assets, capacity):
    # assets: list of tuples (expected_return, risk)
    # capacity: maximum risk we can take

    # compute return-to-risk ratio
    assets = [(er, risk, er/risk) for er, risk in assets]

    # sort by ratio descending
    assets.sort(key=lambda x: x[2], reverse=True)

    remaining = capacity
    total_return = 0
    allocation = []

    for er, risk, ratio in assets:
        if risk <= remaining:
            # take full asset allocation
            allocation.append((er, risk, 1.0))
            total_return += er
            remaining -= risk
        else:
            # take fraction
            fraction = remaining / risk
            allocation.append((er, risk, fraction))
            total_return += er * fraction
            break

    return allocation, total_return
```
