```python
def countMajoritySubarrays(nums, target):
    A = [1 if x == target else -1 for x in nums]

    transformation = [0]
    for i in A:
        transformation.append(transformation[-1] + i)

    def merge_sort(numbers):
        if len(numbers) <= 1:
            return numbers, 0

        mid = len(numbers) // 2
        left_side = numbers[:mid]
        right_side = numbers[mid:]

        left_side, left_count = merge_sort(left_side)
        right_side, right_count = merge_sort(right_side)

        merge = []
        i = j = 0
        count = left_count + right_count

        while i < len(left_side) and j < len(right_side):
            if left_side[i] < right_side[j]:
                # Count all remaining right_side[j..] since left[i] < right[j]
                count += len(right_side) - j
                merge.append(left_side[i])
                i += 1
            else:
                merge.append(right_side[j])
                j += 1

        merge.extend(left_side[i:])
        merge.extend(right_side[j:])

        return merge, count

    _, result = merge_sort(transformation)
    return result


# Call the function and print the output
print(countMajoritySubarrays(nums=[1, 2, 2, 3], target=2))
```
