```python
def find_password(input_text):

    with open(input_text) as f:
        data = f.read().splitlines()
        
        pointing = 50
        count = 0
        for i in data:
            if i[0] == 'L':
                pointing = (pointing - int(i[1:])) % 100
            elif i[0] == 'R':
                pointing = (pointing + int(i[1:])) % 100

            if pointing == 0:
                count += 1
    
        return count

find_password(input_text='input.txt')
```

```python
def is_invalid(num):
    
    s = str(num)
    n = len(s)
    if n % 2 != 0:
        return False
    
    left = s[:n//2]
    right = s[n//2:] # divide and conquer
    return left == right


def invalid_ids(text_data):

    count = 0
    with open(text_data) as f:
        data = f.read().splitlines()
        for x in data:
            list_ids = x.split(',')
            for list_id in list_ids:
                first, second = list_id.split('-')
                for x in range(int(first), int(second) + 1):
                    if is_invalid(x):
                        count += x
    return count    

invalid_ids(text_data='invalid_ids.txt')
```

```python
def repeatedSubstringPattern(s: str) -> bool:

        return s in (s + s)[1:-1]

repeatedSubstringPattern(s='ababab')
```

```python
def findRepeatedDnaSequences(s: str):
    from collections import Counter
    counter = Counter(s[i:i+10] for i in range(len(s) - 9))

    return [seq for seq, count in counter.items() if count > 1]
    
findRepeatedDnaSequences(s="AAAAACCCCCAAAAACCCCCCAAAAAGGGTTT")
```

```python
# advent of code day 3

def find_highest_joltage(data):

    with open(data) as f:
        data = f.read().split()
        
        count = 0
        for d in data:
            highest = 0
            for i, first_highest in enumerate(d[:-1]):
                second_highest = max(d[i+1:])
                combined = int(first_highest + second_highest)
                if combined > highest:
                    highest = combined
            
            count += highest
        return count

find_highest_joltage(data='joltage_bank.txt')
```
