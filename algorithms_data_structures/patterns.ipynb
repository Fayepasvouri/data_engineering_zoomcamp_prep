```python
# given an integer N, print the start pattern rectangle

def print_5_star_pattern(lines, stars):

    return [(stars * '*') for i in range(lines)]

print_5_star_pattern(lines= 8, stars=5)
```

```python
def print_stars_from_1_to_10(length):

    return [(stars * '*') for stars in range(1, length+1)]

print_stars_from_1_to_10(length=10)
```

```python
def print_nums_from_1_to_10(length):

    for i in range(1, length+1):
        pattern = ''.join(str(i) for i in range(1, i+1))
        print(pattern)
        

print_nums_from_1_to_10(length=10)
```

```python
def print_nums_from_1_to_10(length):

    for i in range(1, length+1):
        print(str(i) * i)
        

print_nums_from_1_to_10(length=10)
```

```python
def print_5_star_pattern(lines, stars):

    return [(stars * '*') for i in range(1, lines, 2)]

print_5_star_pattern(lines= 8, stars=5)
```

```python
def print_rectangle_pattern(width, height):

    print('*' * width)

    for i in range(height-2):
        print('*' + ' ' * (width-2) + '*')
    

    print('*' * width)
   

print_rectangle_pattern(5, 4)
```
