#### factorial recursion O(n), fibonacci 0(2^n), binary search (logn)

```python
# leetcode example 231 

# recursively a function calls itself :)

class Solution:
    def isPowerOfTwo(self, n: int) -> bool:
        if n == 1:
            return True
        if n <= 0 or n % 2 != 0:
            return False
         
        return self.isPowerOfTwo(n//2) # O(log2(n)) time complexity cos n is divided by 2 multiple times
```

```python
class Solution:
    def addDigits(self, num: int) -> int:

        if num < 10:
            return num
    
        digit_extraction = [int(digits) for digits in str(num)]
        
        summation = sum(digit_extraction)
        
        return self.addDigits(summation)
        
        # time complexity O(d) where d is number of digits in num

result = Solution()
result.addDigits(38)
# leetcode example 258

# second solution 

def addDigits(num):
    
    if num < 10:
        return num
    
    digit_extraction = str(num).split()
    
    digits = [int(digits) for digits in ",".join(digit_extraction)]
    
    summation = sum(digits)
    
    return addDigits(summation)
        
addDigits(num=38)
```
