```python
# leetcode example 21 merge two sorted linked lists

list1 = [1,2,4]
list2 = [1,3,4]

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
    
def list_to_linkedlist(lst):
    head = None
    for val in reversed(lst):
        head = ListNode(val, head)
    return head
    
list1 = [1,2,4]
list2 = [1,3,4]

dummy = ListNode(0)
current = dummy
p1, p2 = list_to_linkedlist(list1), list_to_linkedlist(list2)

while p1 and p2:
    if p1.val < p2.val:
        current.next = p1
        p1 = p1.next
        current = current.next
    elif p1.val == p2.val:
        current.next = p1
        p1 = p1.next
        current = current.next
                
        current.next = p2
        p2 = p2.next
        current = current.next
    else:
        current.next = p2
        p2 = p2.next
        current = current.next

if p1:
    current.next = p1
else:
    current.next = p2
current = dummy.next
print(current)
```
