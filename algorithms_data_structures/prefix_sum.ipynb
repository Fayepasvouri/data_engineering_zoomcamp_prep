```python
# Prefix sum + Range Sum 

def build_prefix(nums):
    prefix = [0] * len(nums)
    prefix[0] = nums[0]

    for i in range(1, len(nums)):
        prefix[i] = prefix[i - 1] + nums[i]

    return prefix


def range_sum(prefix, l, r):
    if l == 0:
        return prefix[r]
    return prefix[r] - prefix[l - 1]
```

```python
# Prefix sum + hashmap

from collections import defaultdict

def prefix_sum_subarrays(nums, k):

    count = 0
    prefix_sum = 0
    seen = defaultdict(int)
    seen[0] = 1

    for x in nums:
        prefix_sum += x

        if prefix_sum - k in seen:
            count += seen[prefix_sum - k]

        seen[prefix_sum] += 1

    return count
print(prefix_sum_subarrays(nums = [1, 1, 1, 2, 3, 4, -1, -9, 6], k = 2))
```

```python
# Prefix sum + longest subarray k 

def longest_subarray_sum_k(nums, k):
    prefix_sum = 0
    first_seen = {0: -1}
    longest = 0

    for i, x in enumerate(nums):
        prefix_sum += x

        if prefix_sum - k in first_seen:
            length = i - first_seen[prefix_sum - k]
            longest = max(longest, length)

        if prefix_sum not in first_seen:
            first_seen[prefix_sum] = i

    return longest
    return longest
print(longest_subarray_sum_k([1, -1, 5, -2, 3], 3)) 
```
