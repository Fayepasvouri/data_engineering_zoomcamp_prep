```python
def mobius(n):
    if n == 1:
        return 1

    primes = 0
    d = 2
    x = n

    while d * d <= x:
        if x % d == 0:
            primes += 1
            x //= d
            if x % d == 0:
                return 0
        d += 1

    if x > 1:
        primes += 1

    return -1 if primes % 2 == 1 else 1

for i in range(1, 10):
    print(i, mobius(i))
```
