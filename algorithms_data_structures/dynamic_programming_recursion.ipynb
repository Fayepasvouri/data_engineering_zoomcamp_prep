```python
from functools import lru_cache

depth = 0
def pp(f):
    def foo(*args):
        global depth
        print(" |"*depth, *args)
        depth +=1
        res = f(*args)
        depth -=1
        print(" |"*(depth)+" -", res) 
        return res
    return foo
        
# @pp
@lru_cache(None)
def number_paths_rectangle(n, m):
    if n<1 : return 0
    if m<1 : return 0
    if n==1 and m==1:
        return 1
    return number_paths_rectangle(n-1,m) + number_paths_rectangle(n,m-1)
    

print(number_paths_rectangle(200,200))
```

```python
# the recurrence depends on previous 2 values in dynamic programming (climbing stairs and fibonnaci rely on last 2 values)

@lru_cache(None)
def climb_stairs(n):

    if n <= 2:
        return n
    
    dp = [0] * (n + 1)
    dp[1], dp[2] = 1, 2

    for i in range(3, n + 1):
        dp[i] = dp[i-1] + dp[i-2]
        print(dp[i])
    return dp[n]

print(climb_stairs(10)) # its O(n) cos it stores previous steps in memory 
```

```python
# debugging (recursive) code

indent = 0
def print_stack(foo):
    def f(*args, **kwargs):
        global indent
        print(" │"*(indent-1)+" "+"┌", f"{foo.__name__}({','.join(map(repr,args))
        }{kwargs if kwargs else ''})")
        indent += 1
        v = foo(*args, **kwargs)
        indent -= 1
        print(" │"*(indent-1)+" "+"└►", v)
        return v 
    return f

# @print_stack
```
