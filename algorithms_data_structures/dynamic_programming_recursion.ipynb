```python
from functools import lru_cache

depth = 0
def pp(f):
    def foo(*args):
        global depth
        print(" |"*depth, *args)
        depth +=1
        res = f(*args)
        depth -=1
        print(" |"*(depth)+" -", res) 
        return res
    return foo
        
# @pp
@lru_cache(None)
def number_paths_rectangle(n, m):
    if n<1 : return 0
    if m<1 : return 0
    if n==1 and m==1:
        return 1
    return number_paths_rectangle(n-1,m) + number_paths_rectangle(n,m-1)
    

print(number_paths_rectangle(200,200))
```
