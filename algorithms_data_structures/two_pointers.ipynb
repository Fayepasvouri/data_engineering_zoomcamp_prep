### Two pointers Explanation:

1. You start with two pointersâ€”often at the start and end of the array/string.

2. In each iteration, you compare or process the elements at these pointers.

3. Then move one or both pointers based on conditions relevant to the problem.

4. Repeat until pointers meet or cross.

5. O(n) time

```python
from typing import List 


def twoSum(numbers: List[int], target: int) -> List[int]:

    left_pointer = 0
    right_pointer = len(numbers) - 1

    while left_pointer < right_pointer:
        sum_nums = numbers[left_pointer] + numbers[right_pointer]

        if sum_nums == target:
            return [left_pointer + 1,  right_pointer + 1]
        
        elif sum_nums > target:
            right_pointer -= 1
        elif sum_nums < target:
            left_pointer += 1
    
    return []

    
twoSum(numbers = [2,7,11,15], target = 9)
```
