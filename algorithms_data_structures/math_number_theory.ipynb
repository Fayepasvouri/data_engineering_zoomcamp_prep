```python
# 829 leetcode consecutive numbers sum (hard)

# math logic 

# sequence has length of k 

# n = k*x + k(k-1)/2

# solve for x

# x = (n - k * (k-1)/2) / k

# k = length 

# x = positive integer

# n = sum of consequtive numbers


def consecutiveNumbersSum(n: int) -> int:

    k = 1

    count = 0

    while k * (k + 1) // 2 <= n:
        if (n - k * (k-1) // 2) % k == 0:
            count += 1

        k += 1
    
    return count

consecutiveNumbersSum(n = 10)
```

```python
# 2177 leetcode

def sumOfThree(num: int):

    if num % 3 != 0:
        return []

    x = num // 3 - 1
    return [x, x+1, x+2]


sumOfThree(num=33)
```
