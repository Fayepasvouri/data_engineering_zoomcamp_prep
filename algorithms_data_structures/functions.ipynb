```python
# map and why it is used 

lista = [1,2,4,5]

result = list(map(lambda x: x**2, lista))

result 

# map applies a given function to each item of an iterable (like a list) and returns a map object (which is an iterator). 
# It is often used to apply a transformation to all elements in a collection efficiently. Similar to reduce. 
```

```python
# The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, 
# and then the second item in each passed iterator are paired together etc.

a = {"John", "Charles", "Mike"}
b = {"Jenny", "Christy", "Monica", "Vicky"}

x = zip(a, b)

# tuple operates like list with append, accessing with index [0] etc
```

```python
import re

#Split the string at the first white-space character:

txt = "The rain in Spain"
x = re.split(r"\s", txt, 1)
print(x)
```
